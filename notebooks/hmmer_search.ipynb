# HMMER search for all spider genomes using hmm profile

```python
from spider_silkome_module import (
    RAW_DATA_DIR,
    INTERIM_DATA_DIR,
    PROCESSED_DATA_DIR,
    REFERENCES_DIR,
)
from spider_silkome_module import run_cmd
```

## Configuration

```python
task_name = "hmmer_search_20260211"

spider_genome_dir = RAW_DATA_DIR / "01.ref_gff"

# hmmer_search
hmmer_model_dir = REFERENCES_DIR / "2025_Schoneberg_data" / "hmmer_nucl_profile_trimmed"
nhmmer_search_output = INTERIM_DATA_DIR / "automated_spidroin_annotation" / "nhmmer_search"

# parse_hmmer_search_results
parse_nhmmer_search_results_output = PROCESSED_DATA_DIR / task_name / "nhmmer_search_parsed"
```

## Hmmer search

```python
run_cmd(
    f"pixi run python -m spider_silkome_module.run_nhmmer \
        --genome_path {spider_genome_dir} \
        --hmm_dir {hmmer_model_dir} \
        --output_path {nhmmer_search_output} \
        --threads 70 --force True",
    [nhmmer_search_output]
)
```

## Parse hmmer search results

```python
run_cmd(
    f"pixi run python -m spider_silkome_module.parse_nhmmer_results \
        --input-path {nhmmer_search_output} \
        --output-path {parse_nhmmer_search_results_output} \
        --e-value-threshold 1e-10 \
        --coverage-threshold 0.9",
    [parse_nhmmer_search_results_output]
)
```
